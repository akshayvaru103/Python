# Group by in Pandas

### Chance of Tornado happening today is 1 hundreth of a percent

````python
chance_of_tornado = 0.01/100
np.random.binomial(100000, chance_of_tornado)
````

### Chance of Tornado happening two days in a row

````python
chance_of_tornado = 0.01

tornado_events = np.random.binomial(1, chance_of_tornado, 1000000)
    
two_days_in_a_row = 0
for j in range(1,len(tornado_events)-1):
    if tornado_events[j]==1 and tornado_events[j-1]==1:
        two_days_in_a_row+=1

print('{} tornadoes back to back in {} years'.format(two_days_in_a_row, 1000000/365))
````

Formula for standard deviation
$$\sqrt{\frac{1}{N} \sum_{i=1}^N (x_i - \overline{x})^2}$$

## Chi Square Distribution has only 1 parameter - Degrees of Freedom 
### As DOF increase the skewness decreases 

````python
chi_squared_df2 = np.random.chisquare(2, size=10000)
stats.skew(chi_squared_df2)
````

````python
chi_squared_df5 = np.random.chisquare(5, size=10000)
stats.skew(chi_squared_df5)
````

````python
chi_squared_df8 = np.random.chisquare(8, size=10000)
stats.skew(chi_squared_df8)
````

### In the graph below, Note that the Skewness decrease as DOF increases from --> (2-5-8)
### DOF represents the number of sample taken from the population

````python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

output = plt.hist([chi_squared_df2,chi_squared_df5, chi_squared_df8], bins=50, histtype='step', 
                  label=['2 degrees of freedom','5 degrees of freedom','8 degrees of freedom'])
plt.legend(loc='upper right')
````
