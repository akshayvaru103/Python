```python
#### Import all required libraries
import cv2
import numpy as np
np.set_printoptions(threshold=np.nan)
```

####  We can initially load an image from a file and later use the webcam to scan a sodoku image


```python
#### Load our input image
image = cv2.imread('./image_data/sudoku1.jpg',0) ## Load image in grayscale using the 0 parameter
org_image = image
cv2.imshow('Original', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#### Find the edges of the image
# Find Canny edges
edged = cv2.Canny(image, 50, 200)
cv2.imshow('1 - Canny Edges', edged)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Find contours and print how many were found
ret, contours, hierarchy = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print ("Number of contours found = ", len(contours[0]))
print(type(ret))
print(type(contours))
print(type(hierarchy))
```

```python
print(ret)
cv2.imshow('Original', ret)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(contours)
cv2.imshow('Original', contours)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(hierarchy)
```

```python
#### Sort Contours by area - Function

def get_contour_areas(contours):
    # returns the areas of all contours as list
    all_areas = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        all_areas.append(area)
    return all_areas

# Sort contours large to small
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)
c = sorted_contours[0]
print(c)
```

```python
#### We assume the Largest contour is the outer border of our sudoku 
cv2.drawContours(image, [c], -1, (255,0,0), 3)
cv2.imshow('Contours by area', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#### Next lets extract this part of the image alone 
peri = cv2.arcLength(c, True)
approx = cv2.approxPolyDP(c, 0.02 * peri, True)
print(peri)
print(approx)
```

```python
##### Function to order the vertices of the contour as returned by the approx function
##### [ TOP-LEFT, TOP-RIGHT, BOTTOM-RIGHT, BOTTOM-LEFT ]
##### First take the sum of x,y coordinates. TOP-LEFT has least sum, and BOTTOM-RIGHT has maximum sum. 
##### Next find the difference, ie y-x. TOP-RIGHT has minimum and BOTTOM-LEFT has maximum. 
##### Return points as per the new order
def order(h):
        h = h.reshape((4,2))
        hnew = np.zeros((4,2),dtype = np.float32)
        add = h.sum(1)
        hnew[0] = h[np.argmin(add)]
        hnew[2] = h[np.argmax(add)]
        diff = np.diff(h,axis = 1)
        hnew[1] = h[np.argmin(diff)]
        hnew[3] = h[np.argmax(diff)]  
        return hnew
```

```python
##### Use the function created earlier to get the values in order
new_approx = order(approx)
print(approx)
print(new_approx)
```

```python
#### Find distance between 2 points and get the maximum length of a side
import math
def distance(point1, point2):
    dist = pow((point2[0] - point1[0]),2) + pow((point2[1] - point1[1]),2)
    dist = math.sqrt(dist)
    return int(dist)
def get_max_length(points):
    length = 0
    for i in range(0,len(points)):
        if(i == (len(points)-1)):
            len1 = distance(points[i], points[0])
        else:
            len1 = distance(points[i], points[i+1])
        print(len1)
        if(len1 > length):
            length = len1
    return length
```

```python
#### Apllying Warp transform to crop the region of interest
side = get_max_length(new_approx)
new_points = np.float32([[0,0], [side,0], [side,side], [0,side]])
print(new_points)

M = cv2.getPerspectiveTransform(new_approx, new_points)
 
warped = cv2.warpPerspective(org_image, M, (side,side))
 
cv2.imshow('warpPerspective', warped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##### Find horizontal lines in the image
kernel = np.ones((5,5), np.uint8)
Erosion = cv2.erode(warped, kernel, iterations = 1)
cv2.imshow('Erosion', Erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lines = cv2.HoughLinesP(Erosion, 1, np.pi / 45, 200, 5)
print(lines.shape)
print(lines)

for [[x1, y1, x2, y2]] in lines:
    cv2.line(Erosion, (x1, y1), (x2, y2),(0, 255, 0), 3)

cv2.imshow('Probabilistic Hough Lines', Erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
