### Matplotlib calls to the scripting interface, just creates figures, subplots, and axis. It then, loads those axis up with various Artists, which the back-end renders to the screen or some other medium like a file. 

### While you'll spend 95% of your time at the scripting layer, creating graphs & charts, it's important to also understand how the library works underneath for the other 5% of the time.

```python
from matplotlib.artist import Artist

def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        print(" " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

rec_gc(legend)
```

## Histograms
#### Histogram is a bar chart which shows the frequency of a given phenomena. 
### A Probability function can be visualized as a curve, where the y-axis holds the given (x-component) value would occur & the x-axis is the value itself. This is called Probability Density function

```python
import numpy as np
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size = sample_size)
    axs[n].hist(sample, bins = 100)
    axs[n].set_title('n={}'.format(sample_size))
```

 ## Boxplot || Box and Whisker Plots
#### A method of showing aggregate statistics of various samples in a concise manner. A box plot simultaneously shows, for each sample, the median of each value, the minimum & maximum of the samples, and the Interquartile range. 

```python
import pandas as pd

normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal':normal_sample,
              'random':random_sample,
              'gamma' :gamma_sample})
```

```python
df.describe()
```

```python
plt.figure()
_ = plt.boxplot(df['normal'], whis='range')
```

```python
plt.clf()
_ = plt.boxplot([df['normal'],df['random'],df['gamma']], whis='range')
```

```python
plt.figure()
_ = plt.hist(df['gamma'], bins=100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'],df['random'],df['gamma']], whis='range')
ax2 = mpl_il.inset_axes(plt.gca(), width='50%', height='50%', loc=2)
ax2.hist(df['normal'], bins=100)
ax2.margins(x=0.5)
ax2.yaxis.tick_right()
```

```python
# Importance of whis arguement in the box plot function
#by default, the whiskers actually go out to halfway between the interquartile range, unless whis argument is included
plt.figure()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']])
```

## Heat Maps
#### Heat are a way to visulaize 3 Dimentional Data, to take advantage of spatial proximity of those dimensions. But, Heatmaps are powerful only when you have the right data, Weather Data being a great example. It has 2 dimensions Latitude & Longitude & you can overlay on top of this, a 3rd dimension, say, Temperature || Rainfall amounts - and use color to indicate its intensity.  

### Heat Maps don't work when there is no continuous relationships. Don't  use Heat Maps for Categorical Data. It misleads the viewer into looking for patterns and ordering through spatial proximity. 

## In Matplotlib a Heat Map is just a 2D Histogram

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()   # Clearing the figure
    bins = np.arange(-4,4,0.5)  # We can also pass spacing in between bins instead of just 10|100
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_xlabel('Values')
    plt.gca().set_ylabel('Frequency')
    plt.annotate('n = {}'.formate(curr), [3,27])
```
