```python
import cv2
import numpy as np
import face_recognition
```

```python
names = ['muthu','thilaga','sundar','nithin','akshay']
folder = '/home/pracas/Machine Vision/Workshop/workshop photos/test'
```

```python
face_encodings = []

for i in range(0,5):
    file_name = folder +"/"+names[i]+".jpg"
    #print(file_name)
    image = cv2.imread(file_name)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    known_image = cv2.resize(image, (0, 0), fx=0.35, fy=0.35)
    face_locations = face_recognition.face_locations(known_image)
    (t,r,b,l) = face_locations[0]
    face = known_image[t:b,l:r]
    known_encoding = face_recognition.face_encodings(face)[0]
    face_encodings.append(known_encoding)
print(face_encodings)
```

```python
file_name = folder +"/"+"sundaresan_test"+".jpg"
unknown_image = cv2.imread(file_name)
unknown_image = cv2.resize(unknown_image, (0, 0), fx=0.35, fy=0.35)
unknown_image = cv2.cvtColor(unknown_image,cv2.COLOR_BGR2RGB)
face_locations = face_recognition.face_locations(unknown_image)
(t,r,b,l) = face_locations[0]
face = unknown_image[t:b,l:r]
unknown_encoding = face_recognition.face_encodings(face)[0]
#print(unknown_encoding)
face_distances = face_recognition.face_distance(face_encodings,unknown_encoding)
print(face_distances.argmin())
```

```python
file_name = folder +"/"+"VID_20180729_152440452.mp4"
cap = cv2.VideoCapture(file_name)
cv2.namedWindow("video")
cv2.waitKey()
frame_count = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    frame_count = frame_count + 1
    if ret == True :
        if(frame_count%15 == 0):
            frame = cv2.resize(frame, (0, 0), fx=0.4, fy=0.4)
            frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            face_locations = face_recognition.face_locations(frame)
            if(len(face_locations)>0):
                for (t, r, b, l) in face_locations:
                    unknown_encoding = face_recognition.face_encodings(frame, [(t, r, b, l)])[0]
                    face_distances = face_recognition.face_distance(face_encodings,unknown_encoding)
                    print(face_distances)
                    text =""
                    if(np.min(face_distances)<0.4):
                        face_index = np.argmin(face_distances)
                        text = text + names[face_index]
                    cv2.rectangle(frame, (l, t), (r, b), (0, 0, 255), 2)
                    font = cv2.FONT_HERSHEY_COMPLEX_SMALL
                    cv2.putText(frame,text,(l,t), font, 1,(255,0,255),1,cv2.LINE_AA)
            cv2.imshow("video",frame)
            cv2.waitKey(1)
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
